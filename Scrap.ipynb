```python
from bs4 import BeautifulSoup
import requests
html_text=requests.get('https://www.ketabrah.ir/books/%DA%A9%D8%AA%D8%A7%D8%A8-%D8%AF%D8%A7%D9%86%D9%84%D9%88%D8%AF-%DA%A9%D8%AA%D8%A7%D8%A8-%D8%AF%D8%A7%D8%B3%D8%AA%D8%A7%D9%86-%D8%B1%D9%85%D8%A7%D9%86-%D8%AE%D8%A7%D8%B1%D8%AC%DB%8C?bt=books&sort=populars').text
soup=BeautifulSoup(html_text , 'lxml')
Books=soup.find('div',class_="book-list")
Books_item=Books.find_all('div',class_="item")
for Book in Books_item:
    Book_Name=Book.find('a',class_='title').text
    Book_Athour=Book.find('div',class_='authors').text
    print(f"book's name:{Book_Name.strip()}")
    print(f"book's athour:{Book_Athour.strip()}")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage
url = "https://www.ketabrah.ir/books/%DA%A9%D8%AA%D8%A7%D8%A8-%D8%AF%D8%A7%D9%86%D9%84%D9%88%D8%AF-%DA%A9%D8%AA%D8%A7%D8%A8-%D8%AF%D8%A7%D8%B3%D8%AA%D8%A7%D9%86-%D8%B1%D9%85%D8%A7%D9%86-%D8%AE%D8%A7%D8%B1%D8%AC%DB%8C?bt=books&sort=populars"

# Send a GET request
response = requests.get(url)

# Parse the webpage content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all book sections
books = soup.find_all('a', class_='cover')

# Extract and print information
for book in books:
    title = book['title']
    link = "https://www.ketabrah.ir" + book['href']
    img_src = book.find('img')['src']
    discount_label = book.find('div', class_='cover-discount-label').text if book.find('div', class_='cover-discount-label') else "No discount"
    
    print(f"Title: {title}")
    print(f"Link: {link}")
    print(f"Image: {img_src}")
    print(f"Discount: {discount_label}")
    print("-" * 40)
```
